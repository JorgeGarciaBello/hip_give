## Función para preparar los datos de texto

```python
def text_analysis(serie):
    s=serie.isna().value_counts()    
    if s.size==2:
          serie=serie.dropna()
    elif (s.size==1)&(s[0]==True):
      print('no existen datos')
    datetime_object = datetime.datetime.now()
    lista=[]    
    for u_text in serie:
        if(type(u_text)!= type(datetime_object)):            
            lista.append(u_text.encode("utf-8"))
    text_string_type=', '.join(lista)
    text_string_type=unicode(text_string_type, "utf-8").lower()    
    
    useless_words=nltk.corpus.stopwords.words("spanish")+nltk.corpus.stopwords.words("english")+list(string.punctuation)+[u'cada',u'través',u'para',u'\xa1que',u'\xa1to',u'son',u'con',u'la',u'se',u'en',u'el',u'que',u'la',u'las',u'lo',u'los',u'de',u'the',u'us',u'we',u'A',u'nbsp',u'8217',u'8211',u'8220',u'8221',u'\u2019',u'\u201d',u'\u201c',u'pa\xeds',u'visi\xf3n',u'ray\xf3n',u'pa\xeds',u'tecnolog\xeda',u'ay\xfadanos',u'\xa1que',u'harder\u2014but',u'michoac\xe1n',u'let\xb4s',u'\xa1to',u'granny\xb4s',u'espa\xf1ol.',u'.']#+list(u"\u2019")+list('...')    
    words=nltk.word_tokenize(text_string_type)
    words_filtered=[word for word in words if not word in useless_words]
    
    words_counter=Counter(words_filtered)
    most_common_words = words_counter.most_common()
    return most_common_words
```

## Análisis de comentarios

```python
message_words=text_analysis(message_crean)
```

```python
message_words
```

```python
# hpgraph es una función transforma una lista en una tupla con dos listas para 
# poder crear gráfias
def hpgraph(serie):
    serie=serie[:10]
    x=[]
    y=[]
    
    for s in serie:        
        x.append(s[0])
        #y.append(float(s[1])/float(total_rows))
        y.append(float(s[1]))
    return (x,y)
```

```python
data_to_graph=hpgraph(message_words)
```

```python
plt.plot(data_to_graph[0], data_to_graph[1], 'r--')
xTicks=data_to_graph[0]
plt.title(u'Palabras más utilizadas \n en los mensajes de los donantes')
plt.ylabel(u'# de eventos')
plt.xticks(data_to_graph[0], xTicks)
plt.xticks(range(10), xTicks, rotation=90) #writes strings with 90 degree angle
plt.savefig('words_by_donors.png',bbox_inches='tight',dpi=150)
```

## Importando conjunto de datos

```python
data_source_url = "https://raw.githubusercontent.com/kolaveridi/kaggle-Twitter-US-Airline-Sentiment-/master/Tweets.csv"
#data_source = "data/data_for_training_sentimental_analysis.csv"
airline_tweets = pd.read_csv(data_source_url)
```

```python
airline_tweets.head(2)
```

## Limpieza de datos

```python
features = airline_tweets.iloc[:, 10].values
labels = airline_tweets.iloc[:, 1].values
```

```python
type(features[0])
```

```python
labels[0:2]
```

```python
processed_features = []

for sentence in range(0, len(features)):
    # Remove all the special characters
    processed_feature = re.sub(r'\W', ' ', str(features[sentence]))

    # remove all single characters
    processed_feature= re.sub(r'\s+[a-zA-Z]\s+', ' ', processed_feature)

    # Remove single characters from the start
    processed_feature = re.sub(r'\^[a-zA-Z]\s+', ' ', processed_feature) 

    # Substituting multiple spaces with single space
    processed_feature = re.sub(r'\s+', ' ', processed_feature, flags=re.I)

    # Removing prefixed 'b'
    processed_feature = re.sub(r'^b\s+', '', processed_feature)

    # Converting to Lowercase
    processed_feature = processed_feature.lower()

    processed_features.append(processed_feature)
```

## Representando texto en forma numérica

```python
# Técnica "TF-IDF"
vectorizer = TfidfVectorizer (max_features=2500, min_df=7, max_df=0.8, stop_words=stopwords.words('english')+stopwords.words('spanish'))
processed_features = vectorizer.fit_transform(processed_features).toarray()
```

```python
processed_features[0].size  # processed_features es un numpy.array
```

## Dividiendo datos en un conjunto para entrenar y uno para realizar pruebas

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(processed_features, labels, test_size=0.20, random_state=0)
```

## Entrenando el modelo


```python
from sklearn.ensemble import RandomForestClassifier

text_classifier = RandomForestClassifier(n_estimators=200, random_state=0)
text_classifier.fit(X_train, y_train)
```

## Evaluación del modelo en datos de prueba

```python
predictions = text_classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

#print(confusion_matrix(y_test,predictions))
print(classification_report(y_test,predictions))
print(accuracy_score(y_test, predictions))
print(accuracy_score(y_test, predictions))
```
