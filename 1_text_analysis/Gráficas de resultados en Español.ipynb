```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
short_description_words_cero_donation=pd.read_excel('data/short_description_words_cero_ESP.xlsx')
short_description_words_goal_donation=pd.read_excel('data/short_description_words_goal_ESP.xlsx')

long_desciption_words_cero_donation=pd.read_excel('data/long_description_words_cero_ESP.xlsx')
long_desciption_words_goal_donation=pd.read_excel('data/long_description_words_goal_ESP.xlsx')
```

```python
# hpgraph es una función transforma una lista en una tupla con dos listas para 
# poder crear gráfias
def hpgraph(serie, total_rows):
    serie=serie[:10]
    x=[]
    y=[]
    
    for s in serie:        
        x.append(s[0])
        y.append(float(s[1])/float(total_rows))    
    return (x,y)
```

```python
data_for_graph_of_short_cero_donation=hpgraph(short_description_words_cero_donation.values.tolist(),len(short_description_words_cero_donation.values.tolist()))
data_for_graph_of_short_goal_donation=hpgraph(short_description_words_goal_donation.values.tolist(),len(short_description_words_goal_donation.values.tolist()))
```

```python
data_for_graph_of_long_cero_donation=hpgraph(long_desciption_words_cero_donation.values.tolist(),len(long_desciption_words_cero_donation.values.tolist()))
data_for_graph_of_long_goal_donation=hpgraph(long_desciption_words_goal_donation.values.tolist(),len(long_desciption_words_goal_donation.values.tolist()))
```

## Gráficas de las palabras en Español más usadas en las short description

```python
plt.plot(data_for_graph_of_short_cero_donation[0], data_for_graph_of_short_cero_donation[1], 'r--')
xTicks=data_for_graph_of_short_cero_donation[0]
plt.title(u'Palabras en Español más utilizadas \n en proyectos que no recaudan')
plt.ylabel(u'% ponderado por fila')
plt.xticks(data_for_graph_of_short_cero_donation[0], xTicks)
plt.xticks(range(10), xTicks, rotation=90) #writes strings with 45 degree angle
plt.savefig('spanish_chart_words_cero.png',bbox_inches='tight',dpi=150)
```

```python
plt.plot(data_for_graph_of_short_goal_donation[0], data_for_graph_of_short_goal_donation[1], 'r--')
xTicks=data_for_graph_of_short_goal_donation[0]
plt.title(u'Palabras en Español más utilizadas \n en proyectos que recaudan el objetivo o más')
plt.ylabel(u'% ponderado por fila')
plt.xticks(data_for_graph_of_short_goal_donation[0], xTicks)
plt.xticks(range(10), xTicks, rotation=90) #writes strings with 45 degree angle
plt.savefig('spanish_chart_words_goal.png',bbox_inches='tight',dpi=150)
```

```python
plt.plot(data_for_graph_of_long_cero_donation[0], data_for_graph_of_long_cero_donation[1], 'r--')
xTicks=data_for_graph_of_long_cero_donation[0]
plt.title(u'Palabras en Español más utilizadas en long description \n de proyectos que no recaudan')
plt.ylabel(u'% ponderado por fila')
plt.xticks(data_for_graph_of_long_cero_donation[0], xTicks)
plt.xticks(range(10), xTicks, rotation=90) #writes strings with 45 degree angle
plt.savefig('spanish_chart_long_desc_words_cero.png',bbox_inches='tight',dpi=150)
```

## Unión de gráficas para short description

```python
def all_chart(s_1,s_2,s_3):
    elements1=len(s_1)
    elements2=len(s_2)
    elements3=len(s_3)
    s1=s_1[0:15]
    s2=s_2[0:15]
    s3=s_3[0:15]
    x=[]
    y_1=[]
    y_2=[]
    y_3=[]
    for s_1 in s1:
        x.append(s_1[0])
    for s_2 in s2:
        x.append(s_2[0])
    for s_3 in s3:
        x.append(s_3[0])
    x=list(set(x))
    x.sort()
    
    s1_dict=dict(s1[0:15])
    s2_dict=dict(s2[0:15])
    s3_dict=dict(s3[0:15])
    
    s1_keys=s1_dict.keys()
    s2_keys=s2_dict.keys()
    s3_keys=s3_dict.keys()
    
    for word in x:        
        if word in s1_keys:            
            y_1.append(float(s1_dict[word])/float(elements1))            
        else:
            y_1.append(0)            
        if word in s2_keys:            
            y_2.append((float(s2_dict[word])/float(elements2)))            
        else:
            y_2.append(0)            
        if word in s3_keys:            
            y_3.append((float(s3_dict[word])/float(elements3)))            
        else:
            y_3.append(0)        
    return (x,y_1,y_2,y_3)
```

```python
bag_of_words_short_des_for_chart=all_chart(short_description_words_cero_donation.values.tolist(),[],short_description_words_goal_donation.values.tolist())
```

```python
len(bag_of_words_short_des_for_chart[0])
```

```python
x=bag_of_words_short_des_for_chart[0]
plt.plot(bag_of_words_short_des_for_chart[0],bag_of_words_short_des_for_chart[1],'r--',bag_of_words_short_des_for_chart[3],'b--')
xTicks=x
plt.title(u'Resultados globales\n de las palabras en Español empleadas en la descipción corta')
plt.ylabel(u'% de uso ponderado por fila')
plt.xticks(x, xTicks)
plt.xticks(range(len(bag_of_words_short_des_for_chart[0])), xTicks, rotation=90) #writes strings with 45 degree angle
plt.savefig('chart_words_total_Spanish_short.png',bbox_inches='tight',dpi=150)
```
