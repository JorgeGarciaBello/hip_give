### Gráficos de pastel

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Gráfico de sentimientos

```python
fig, ax = plt.subplots(figsize=(9, 5), subplot_kw=dict(aspect="equal"))

recipe = ["3432  positivo",
          "85  negativo",
          "1808 neutral"
          ]

data = [float(x.split()[0]) for x in recipe]
ingredients = [x.split()[-1] for x in recipe]


def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.2f}%\n({:d})".format(pct, absolute)


wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"))

ax.legend(wedges, ingredients,
          title="Sentimiento",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=13.5, weight="bold")

ax.set_title("Porcentaje de sentimientos \n Base A")
#plt.savefig('pie_chart_comment_analysis_base_A.png',bbox_inches='tight',dpi=150)
#plt.savefig('pie_chart_comment_analysis_base_A.png')
plt.savefig('pie_chart_comment_analysis_base_AA.png',bbox_inches='tight')
```

```python
fig, ax = plt.subplots(figsize=(9, 5), subplot_kw=dict(aspect="equal"))
recipe = ["5260  positivo",          
          "65 neutral"]
data = [float(x.split()[0]) for x in recipe]
ingredients = [x.split()[-1] for x in recipe]
def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d})".format(pct, absolute)
explode = (0.3, 0.0)  # only "explode" the 1-first slice (i.e. 'positivo
wedges, texts, autotexts = ax.pie(data, explode=explode,autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="black"),colors=['tab:blue','tab:green'])
ax.legend(wedges, ingredients,
          title="Sentimiento",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))
plt.setp(autotexts, size=14, weight="bold")
ax.set_title("Porcentaje de sentimientos \n Base B")
plt.savefig('pie_chart_comment_analysis_base_B.png',bbox_inches='tight')
```

## Gráfico de género de donantes

```python
fig, ax = plt.subplots(figsize=(9, 5), subplot_kw=dict(aspect="equal"))

recipe = ["5891  Mujeres",
          "2630  Hombres",
          "2005  Ambiguo",
          "1400  Otros"
          ]
data = [float(x.split()[0]) for x in recipe]
ingredients = [x.split()[-1] for x in recipe]
def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))+1
    print(type(absolute))
    #absolute = int(pct/100.*np.sum(allvals))
    return "{:.2f}%\n({:d})".format(pct, absolute)
wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"))
ax.legend(wedges, ingredients,
          title="Genero",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))
plt.setp(autotexts, size=13.0, weight="bold")

ax.set_title("Naturaleza de los donantes")
plt.savefig('pie_chart_gender_analysis.png',bbox_inches='tight')
```
