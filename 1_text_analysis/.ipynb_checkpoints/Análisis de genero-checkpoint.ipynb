#### Lectura de base de datos

```python
# Lectura de base de datos y obtenci√≥n de nombre del donante
data = pd.read_excel('../STANFORD_ HIPGive_By_country_2014_2019_JUNUETH.xlsx',sheet_name='All funders data 2014-2019')
```

```python
payer_name = data[u'Payer Name']
```

```python
# Limpieza de datos nulos
payer_name = payer_name.dropna()
```

```python
payer_name = payer_name.tolist()
```

```python
#payer_name
```

#### Primer limpieza de nombres

```python
clean_names=[]

for name in range(0, len(payer_name)):
    # Remove all the special characters
    clean_name = re.sub(r'\W', ' ', payer_name[name])

    # remove all single characters
    clean_name= re.sub(r'\s+[a-zA-Z]\s+', ' ', clean_name)

    # Remove single characters from the start
    clean_name = re.sub(r'\^[a-zA-Z]\s+', ' ', clean_name) 

    # Substituting multiple spaces with single space
    clean_name = re.sub(r'\s+', ' ', clean_name, flags=re.I)

    # Removing prefixed 'b'
    clean_name = re.sub(r'^b\s+', '', clean_name)

    # Converting to Lowercase
    clean_name = clean_name.lower()

    clean_names.append(clean_name)
```

```python
len(clean_names)
```

```python
clean_names_set = set(clean_names)
```

```python
len(clean_names_set)
```

```python
clean_names_list = list(clean_names_set)
```

#### Tokenizando nombres

```python
tokenized_names=[]
for name in clean_names_list:
    tokenized_name = nltk.word_tokenize(name)
    tokenized_names.append(tokenized_name)
```

### Refinamiento de an√°lisis, basado en base de datos elaborada por Clave

```python
# Lectura de la base de datos de Clave
data_cl=pd.read_excel('data/Clasificacion_de_donantes.xlsx',sheet_name='Hoja2')
```

```python
clasifications=data_cl['Hombre/Mujer/Ambiguo/Institucion/Otro'].unique()
```

```python
clasifications
```

```python
clasification_values=data_cl['Hombre/Mujer/Ambiguo/Institucion/Otro'].values
```

```python
hombres=[]
mujeres=[]
ambiguo=[]
inst=[]
otro=[]
for clas_val in clasification_values:
    if clas_val == u'Hombre':
        hombres.append(1)
    elif clas_val == u'Mujer ' or clas_val == u'Mujer':
        mujeres.append(1)
    elif clas_val == u'Ambiguo ':
        ambiguo.append(1)
    elif clas_val == u'Instituci\xf3n':
        inst.append(1)
    elif clas_val == u'Otro':
        otro.append(1)
    
```

```python
[['hombres',len(hombres)], ['mujeres',len(mujeres)], ['ambiguo',len(ambiguo)], ['inst',len(inst)], ['otro',len(otro)]]
```
