#### Palabras a estudiar en Español para la descripción corta

```python
spanish_words_in_short_description=[u'acceso',u'apoyar',u'ciudad',u'inmigrantes',u'latinos',u'apoyo',u'estudiantes',u'niñas',u'personas',u'salud',u'violencia',u'comunidades',u'mujeres',u'niños',u'jóvenes']
```

```python
english_words_in_short_description=[u'communities',u'kids',u'latino',u'mexico',u'provide',u'make',u'new',u'school',u'violence',u'people',u'students',u'support',u'woman',u'children']
```

```python
def func_flags(word,_sdEsp,_ids):
    # Creación de lista con datos => [(id,description), ...]
    lista=[]
    if len(_ids)==len(_sdEsp):
        for i in range(0,len(_ids)):
            lista.append((_ids[i],_sdEsp[i]))

    final_list=[]
    for i in range(0,len(_ids)):    
        if word in lista[i][1]:
            final_list.append((_ids[i],1))
        else:
            final_list.append((_ids[i],0))
    return (word,final_list)
```

```python
def func_data_flags(lista_of_words,shortDesc,_ids):
    lista_of_results=[]
    #for i in range(0,len(lista_of_words)):
    #    result=func_flags(lista_of_words[i],shortDesc,_ids)
    #    lista_of_results.append(result)
    if len(_ids)==len(shortDesc):
        for i in range(0,len(_ids)):
            
            lista_per_desc=[]            
            for j in range(0,len(lista_of_words)):
                if  lista_of_words[j] in shortDesc[i]:
                    lista_per_desc.append(1)
                else:
                    lista_per_desc.append(0)
            
            
            #result=func_flags(lista_of_words[i],shortDesc,_ids)
            lista_of_results.append(lista_per_desc)        
    return lista_of_results
```

### Resultados de las banderas en las descripciones cortas en Español

```python
result_flags_shotDescEsp=func_data_flags(spanish_words_in_short_description,_sdEsp,_ids_shorEspData)
```

```python
# Pasando datos a banderas en un ps.DataFrame, para finalmente importarlo a un excel.
flags_short_spanish=pd.DataFrame(result_flags_shotDescEsp, index=_ids_shorEspData, columns=spanish_words_in_short_description)
```

```python
# Coinciden las descripciones si en shape se tiene 736
flags_short_spanish.shape
```

```python
flags_short_spanish.to_excel('data/flags_short_description_spanish.xlsx')
```

### Resultados de las banderas en las descripciones cortas en Inglés

```python
result_flags_shotDescEng=func_data_flags(english_words_in_short_description,_sdEng,_ids_shorEngData)
```

```python
# Pasando datos a banderas en un ps.DataFrame, para finalmente importarlo a un excel.
flags_short_english=pd.DataFrame(result_flags_shotDescEng, index=_ids_shorEngData, columns=english_words_in_short_description)
```
